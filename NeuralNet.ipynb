# Forward Propagation for OR Gate Neural Network using a basic matrix (Building from Scratch

```python
import numpy as np

# The below are the values for Input data, Weights and Bias (X, H1, H2 and B1, B2 respectively)
X = np.array([1, 0])
H1 = np.array([[1, 0, 1, 0], [0, 1, 0, 1]])
B1 = np.array([0.15, 0.15, 0.15, 0.15])
H2 = np.array([1, 1, 0, 1.5])
B2 = np.array([0])

# the activation function is given below
def tanh(x):
    # Or we can also do z = (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))
    return np.tanh(x)

# The below is the calculation of the forward propagation of Hidden Layer which follows the formula Z1 = X . H1 + B1
Z1 = X.dot(H1) + B1
display(Z1)

# I have used the activation function for hidden layer as well as the output layer
A1 = tanh(Z1)
display(A1)

# This is the calculation for the output layer, here I use the output of the activation function in the previous layer Z2 = A1 . H2 + B2 
Z2 = A1.dot(H2) + B2
display(Z2)

# Then using activation on the output of Z2
A2 = tanh(Z2)
display(A2)

print("Final Output (A2):", A2)
```

### Gradient Descent Update for $ H_2 $ and $ B_2 $

```python
import numpy as np

# The below is the derivative of tanh which I will be using for the back propagation
def tderivative(x):
    return 1 - np.tanh(x)**2

# The target value Y is 1 and alpha here is the learning rate (eta)
Y = np.array([[1]])
alpha = 10

# Computing the error gradient for the output using a derivative of MSE (Mean Squared Error)
dL_dA2 = 2 * (A2 - Y)
display(dL_dA2)

# Then calculating the derivative of the output layer (which is Z2) because we used tanh to get y. So in backprop we do the derivative of tanh
dA2_dZ2 = tderivative(Z2)
display(dA2_dZ2)

# Then computing the gradient of Z2 using chain rule
dL_dZ2 = dL_dA2 * dA2_dZ2
display(dL_dZ2)

# Now the gradient of weights for the output layer is calculated by dL_dZ2 . A1 (Changing the shape of A1 to match and using Transpose on whole)
dL_dH2 = dL_dZ2.dot(A1.reshape(1, -1)).T
display(dL_dH2)

# The gradient of B2 is the Gradient of Z2 as their gradient is simply the accumulated gradient from the next layer
dL_dB2 = dL_dZ2.copy()
display(dL_dB2)

# Now I calculate the new weights and bias of the output layer using the gradient and the learning rate which I subract from the old H2 and B2
H2new = H2 - alpha * dL_dH2.T
B2new = B2 - alpha * dL_dB2

print("Updated H2:", H2new)
print("Updated B2:", B2new)
```

# Coding a Neural Network from Scratch

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import multivariate_normal
from nnet import FCLayer, ActivationLayer, Network, sigmoid, sigderivative, tanh, tanhderivative, mse, msederivative

# Initialized Seed value for reproducibility
np.random.seed(42)

# The function tests using the values from the above (code) and checks whether the forward and backward predictions are working or not
def testing():
    X = np.array([[1, 0]])

    H1 = np.array([[1, 0, 1, 0], [0, 1, 0, 1]])
    B1 = np.array([[0.15, 0.15, 0.15, 0.15]])

    H2 = np.array([[1], [1], [0], [1.5]])
    B2 = np.array([[0]])
    
    # Calculating the output through manual forward propagation
    Z1 = X.dot(H1) + B1
    A1 = tanh(Z1)
    Z2 = A1.dot(H2) + B2
    A2 = tanh(Z2)

    # Created the network with one hidden layer (2, 4) size and output layer (4, 1) size and used tanh function as activation
    network = Network()
    network.add(FCLayer(2, 4))
    network.add(ActivationLayer(tanh, tanhderivative))
    network.add(FCLayer(4, 1)) 
    network.add(ActivationLayer(tanh, tanhderivative))

    # Manually assigned values for the Network
    network.layers[0].weights = H1
    network.layers[0].biases = B1
    network.layers[2].weights = H2
    network.layers[2].biases = B2

    # Predicting the output through NN and checking for a match with manual and NN predicted values
    output = network.predict(X)
    assert np.allclose(output, A2, atol=1e-6), "Final output A2 does not match expected value!"
    print("NN Ouput:", output)
    print("Manual Output:", A2)
    print("Forward propagation test passed!")

    # Given target values and learning rate which is represented as alpha here
    Y = np.array([[1]]) 
    alpha = 10

    # Calculating the gradient of the loss function (MSE) and the gradients of each term for back propagation
    dL_dA2 = 2 * (A2 - Y)
    dA2_dZ2 = tderivative(Z2)
    dL_dZ2 = dL_dA2 * dA2_dZ2
    dL_dH2 = A1.T.dot(dL_dZ2)
    dL_dB2 = np.sum(dL_dZ2, axis=0, keepdims=True)

    # I first fit the network with the data (as both forward and backward propagation are done here and the loss is also computed)
    network.fit(X, Y, X, Y, epochs=1, lr=alpha)

    # Then I access the weights and bias of the network for the output layer
    H2updated = network.layers[2].weights
    B2updated = network.layers[2].biases

    # Manually calculating the new H2 and B2
    H2new = H2 - alpha * dL_dH2
    B2new = B2 - alpha * dL_dB2

    # Asserting whether there is a match or not
    assert np.allclose(H2updated, H2new, atol=1e-6), "H2 update mismatch!"
    assert np.allclose(B2updated, B2new, atol=1e-6), "B2 update mismatch!"
    print("Neural Network Calculated")
    print(H2updated)
    print(B2updated)
    print("Manually Calculated")
    print(H2new)
    print(B2new)
    
    print("Backward propagation test passed!")

    print("All tests passed successfully!\n")

testing()

# Now testing the XOR data with the NN architecture
def XorTest():
    # Training Data for XOR
    Xtrain = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
    Ytrain = np.array([[0], [1], [1], [0]])

    # Validation Data for XOR (I changed the values a little to see how it performs)
    Xval = np.array([[0.01, -0.02], [0.02, 1.01], [1.01, -0.01], [0.99, 1.02]])
    Yval = np.array([[0], [1], [1], [0]])

    # Creating a network with tanh as the activation function for both hidden and output layer with sizes (2, 4) and (4, 1)
    network = Network()
    network.add(FCLayer(2, 4))
    network.add(ActivationLayer(tanh, tanhderivative))
    network.add(FCLayer(4, 1))
    network.add(ActivationLayer(tanh, tanhderivative))

    # Fitting the NN with the data, I have used 5000 epoches to see how it works and a learning rate lr of 0.1
    trainLoss, valLoss = network.fit(Xtrain, Ytrain, Xval, Yval, epochs=5000, lr=0.1)

    # Plotting the graph for Training and Validation Loss
    plt.plot(trainLoss, label="Training Loss")
    plt.plot(valLoss, label="Validation Loss", linestyle="dashed")
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('XOR Training vs. Validation Loss Curve')
    plt.legend()
    plt.show()

    # Testing with predictions and changing them to binary
    Ypred = network.predict(Xval)
    YpredB = (Ypred > 0.5).astype(int)
    accuracy = np.mean(YpredB == Yval)
    print(f"Validation Accuracy: {accuracy * 100:.2f}%")

XorTest()

def StressTest():
    # This function is for testing my network against the stress classification data
    # First creating the data using the values for mean and cov
    mstress = np.array([120, 35])
    cstress = np.array([[1, 0], [0, 1]])
    
    mrest = np.array([60, 50])
    crest = np.array([[1, 1], [1, 1]])
    
    # Then using multivariate normal function I created a distribution
    Xstress = multivariate_normal(mstress, cstress, 1000)
    Xrest = multivariate_normal(mrest, crest, 1000)

    # Then creating the same amount of Y values in the dataset
    Ystress = np.ones((1000, 1))
    Yrest = np.zeros((1000, 1))

    # Then stacking both X stress and X rest (I have written "not stress" as "rest")
    X = np.vstack((Xstress, Xrest))
    Y = np.vstack((Ystress, Yrest))

    # Shuffling the data to generalize better
    indices = np.arange(X.shape[0])
    np.random.shuffle(indices)  
    
    X, Y = X[indices], Y[indices]

    # Splitting with the ratio of 0.8 to 0.2
    splitRatio = 0.8
    splitIndex = int(X.shape[0] * splitRatio)
    
    Xtrain, Xval = X[:splitIndex], X[splitIndex:]
    Ytrain, Yval = Y[:splitIndex], Y[splitIndex:]

    # Creating the network with tanh as the activation for hidden layer and sigmoid as the activation for output layer
    network = Network()
    network.add(FCLayer(2, 6))
    network.add(ActivationLayer(tanh, tanhderivative))
    network.add(FCLayer(6, 1))
    network.add(ActivationLayer(sigmoid, sigderivative))

    # Here I am fitting the data and specifying the loss function as well
    losses, valLoss = network.fit(Xtrain, Ytrain, Xval, Yval, epochs=5000, lr=0.01, lossfun=[mse, msederivative])

    # Plotting the graph for Train and Validation loss
    plt.plot(losses, label='Training Loss')
    plt.plot(valLoss, label='Validation Loss', linestyle='dashed')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Training vs Validation Loss Curve')
    plt.legend()
    plt.show()

    # Testing the network with a few samples
    test = np.array([[119.49, 35.15], [120.65, 34.82], [120.35, 35.73], [60.22, 50.31], [59.85, 49.70]])
    predictions = network.predict(test)
    Bpred = (predictions > 0.5).astype(int)
    for i, sample in enumerate(test):
        print(f"Input: {sample}, Prediction: {predictions[i][0]:.4f} ({'Stress' if Bpred[i][0] == 1 else 'Not Stress'})")

StressTest()

# Now I am testing each class (FCLayer and ActivationLayer) seperately using the values

def FCTest():
    # Manually setting up values for testing
    X = np.array([[1, 0]])
    H1 = np.array([[1, 0, 1, 0], [0, 1, 0, 1]])
    B1 = np.array([[0.15, 0.15, 0.15, 0.15]])

    # Calculating the output through manual forward propagation but here I keeping it before activation as I have to test just the FCLayer
    Z1 = X.dot(H1) + B1
    print(f"Z1 (Before Activation): {Z1}")
    
    # Now I use the FCLayer to check if the forward prop is correct or not
    fc = FCLayer(2, 4)
    fc.weights = H1
    fc.biases = B1
    fcOutput = fc.forward_propagation(X)
    
    # Asserting whether they match or not
    assert np.allclose(fcOutput, Z1, atol=1e-6), "FCLayer forward propagation mismatch!"
    print("FCLayer forward propagation passed!\n")

    # Now back propagation with a sample output error that I chose
    operr = np.array([[0.1, -0.1, 0.05, -0.05]])
    fc.backward_propagation(operr, lr=0.01)
    H1new = H1 - 0.01 * np.dot(X.T, operr)
    B1new = B1 - 0.01 * np.sum(operr, axis=0, keepdims=True)
    print(f"Updated Weights after Backpropagation: {fc.weights}")
    print(f"Updated Biases after Backpropagation: {fc.biases}")
    # Checking whether they are correct or not
    print(H1new)
    assert np.allclose(H1new, fc.weights, atol=1e-6), "FCLayer back propagation mismatch!"
    print("FCLayer backward propagation passed!\n")

def ActivationTest():
    # Taking a sample input for activation layer as well
    X = np.array([[1, 0], [0, 1]])
    
    # Now I will be calculating the output using activation function (manually)
    # Expected value for A
    Aexp = tanh(X)

    # Now I am creating an activation layer and testing that specific layer
    actlayer = ActivationLayer(tanh, tanhderivative)
    actop = actlayer.forward_propagation(X)
    
    # Checking if forward propagation is working correctly
    assert np.allclose(actop, Aexp, atol=1e-6), "ActivationLayer forwardprop mismatch!"
    print("ActivationLayer forward propagation passed!\n")

    # Same as before I am randomly selecting an error to test this layer
    operr = np.array([[0.05, -0.02], [0.1, 0.05]])
    nnbackOp = actlayer.backward_propagation(operr, lr=0.01)
    # Checking whether this is correct or not
    backOp = tderivative(X) * operr
    assert np.allclose(nnbackOp, backOp, atol=1e-6), "FCLayer back propagation mismatch!"
    print("ActivationLayer backward propagation completed!\n")

FCTest()
ActivationTest()
```
